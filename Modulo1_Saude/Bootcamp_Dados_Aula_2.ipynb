```python
import pandas as pd
```

```python
dados = pd.read_csv("https://github.com/ErickOlivetto/bootcamp-data-science-aplicada/raw/main/Dados/A233703189_28_143_208.csv", encoding='ISO-8859-1', skiprows=3, sep = ';', skipfooter=12, thousands='.', decimal=',', engine='python')
dados.head()
```

```python
dados.tail()
```

```python
dados.info()
```

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
dados.mean()
```

```python
dados["2008/Ago"]
```

```python
dados["2008/Ago"].mean()
```

```python
dados.plot(x = 'Unidade da Federação', y='2008/Ago')
```

```python
dados.plot(x = 'Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(12 , 8))
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
ax = dados.plot(x = 'Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(12 , 8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.title("Valor por unidade da federação")
plt.show()
```

## Desafio: Escolher um título mais descritivo, que passe a mensagem adequada.

```python
ax = dados.plot(x = 'Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(12 , 8))
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.2f}'))
plt.title("Valor total de processamento para o mês de Agosto de 2008 por unidade da federação")
plt.show()
```

## Desafio: Faça a mesma análise para o mês mais recente.

#Aula 2

```python
dados2 = pd.read_csv("https://github.com/ErickOlivetto/bootcamp-data-science-aplicada/raw/main/Dados/A120703189_28_143_208.csv", encoding='ISO-8859-1', skiprows=3, sep = ';', skipfooter=12, thousands='.', decimal=',', engine='python')
dados2.head()
```

```python
dados2.tail()
```

```python
dados2.info()
```

```python
dados2.plot(x = 'Unidade da Federação', y='2008/Ago', kind = 'bar', figsize=(12 , 8))
```

```python
dados2[['2008/Ago', '2008/Set']]
```

```python
dados2.mean()
```

```python
usaveis = dados2.mean().index.tolist()
usaveis.insert(0, 'Unidade da Federação')
usaveis
```

```python
dados_usaveis = dados[usaveis]
dados_usaveis.head()
```

```python
dados_usaveis = dados_usaveis.set_index('Unidade da Federação')
dados_usaveis.head()
```

```python
dados_usaveis.plot(figsize=(12,8))
```

```python
dados_usaveis.T.plot(figsize=(12,8))
```

```python
dados_usaveis = dados_usaveis.drop('Total', axis=1)
dados_usaveis.head()
```

```python
dados_usaveis.T.plot(figsize=(12,8))
```

## Desafio: Reposicionar a legenda do gráfico em uma posição mais adequada

```python
dados_usaveis.T.plot(figsize=(12,8), legend = False)
plt.legend(bbox_to_anchor=(1, 1), loc = 'upper left')
```

## Desafio: Plotar gráfico de linha com apenas 5 estados da sua preferência

```python
estados = ['35 São Paulo', '41 Paraná', '23 Ceará', '52 Goiás', '13 Amazonas']
dados_usaveis.T[estados].plot(figsize=(12,8), legend = False)
plt.legend(bbox_to_anchor=(1, 1), loc = 'upper left')
```
